### 1. 변수와 함수

1) 변수 (Variable)

[1] 정의 : 어떤 숫자를 대표하는 기호, 주로 $x$, $y$, $z$ 사용

[2] 종류

- 입력 값을 대표 : 입력 변수 (input variable), 정의역 (domain)
- 출력 값을 대표 : 출력 변수 (output variable, 공역 (range)

2) 함수 (Function)

[1] 정의 - 입력 값을 출력 값으로 바꾸어 출력하는 관계, 주로 $f$, $g$, $h$사용

[2] 표현

$$ y = f(x) $$

```python
# 함수 표현하기 (y = 2x)

def f(X):
    return 2 * x  # 함수 만들기 

x = 10
y = f(x)
print(x, y)
```

### 2. 역함수 (inverse function)

1) 정의 : 어떤 함수의 입/출력 관계와 정 반대의 입/출력 관계를 가지는 함수

2) 표시

$$ y = f(x), \;\;\; \rightarrow \;\;\; x = f^{-1}(y) $$

3) 예시

$$ f_1(x) = x^2 \, \;\;\; f_2(x) = \sqrt{x} $$

### 3. 함수와 그래프

1) matplotlib

- 파이썬에선 matplotlib의 라인 플롯을 이용하여 그림
- 구간의 간격이 너무 크면 그래프가 부정확
- 구간의 간격이 너무 적으면 계산 시간이 증가

```python
# matplotlib 이용하여 함수 그리기 (x^3 -3x^2 + x)
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def f(X) :
    return x**3 - 3 * x**2 + x # 함수 생성

x = np.linspace(-1, 3, 9) # -1, 3 사이 9개의 x값을 생성
y = f(x) # x값에 따른 y값 구하기 (9개)

plt.plot(x, y, "ro-")
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.grid(True)

plt.show()
```

```python
# x값을 좀 더 늘려보자
x = np.linspace(-1, 3, 400) # -1, 3 사이 400개의 x값을 생성
y = f(x) # x값에 따른 y값 구하기 (400개)

plt.plot(x, y, "ro-")
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.grid(True)

plt.show() # 아까보다 훨씬 곡선으로 결과가 나옴
```

### 4. 기울기

1) 정의 : 입력 변수의 변화에 따른 출력변수의 변화를 비율로 표현한 것

2) 수식

[1] 일반적

$$ \dfrac{\text{y의 변화량}}{\text{x의 변화량}} = \dfrac{f(x_2) - f(x)}{x_2 - x} = \dfrac{f(x + dx) - f(x)}{dx} $$

[2] 접선의 기울기

$$ \text{기울기} = \lim_{dx\rightarrow 0} \dfrac{f(x + dx) - f(x)}{dx} $$

```python
# 위의 식에서 x=0, x=1에서 접선의 기울기을 구해보자
plt.plot(x, y)
plt.plot(0, 0, 'ro'); plt.plot(x, x, 'r:')
plt.plot(1, -1, 'go'); plt.plot(x, (3*1**2-6*1+1)*(x-1)-1, 'g--')

plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))
plt.grid(True)

plt.annotate('', xy=(1, 0), xytext=(0, 0), arrowprops=dict(facecolor='black'))
plt.annotate('', xy=(1, 1), xytext=(1, 0), arrowprops=dict(facecolor='black'))

plt.annotate('', xy=(2, -1), xytext=(1, -1), arrowprops=dict(facecolor='black'))
plt.annotate('', xy=(2, -3), xytext=(2, -1), arrowprops=dict(facecolor='black'))

plt.show()
```

### 5. 데이터 분석시 많이 사용되는 함수

**1) 다항식 함수 (ploynomial Function)**

[1] 정의 : 상수항 $c_0$, 일차항 $c_1x$, 이차항 $c_2x^2$, $\cdots$ 등의 거듭제곱 항의 선형 조합으로 이루어진 함수

[2] 수식

$$ f(x) = c_0 + c_1 x + c_2 x^2 + \cdots + c_n x^n $$

```python
# 다항 함수 구하기
def f(x) : 
    return 1 + 2 * x + 3 * x ** 2
```

**2) 지수 함수 (exponential function)**

[1] 정의 : 거듭제곱의 지수를 변수로 하고, 정의역을 실수 전체로 정의하는 초월함수

[2] 수식 : $e$는 오일러수 (약 2.718)

$$ y = e^x\;\;or\;\;\ y = \exp x $$

[3] 특징

- 양수(오일러수)를 거듭제곱하기 때문에 음수가 나올 수 없음
- x = 0 $\rightarrow$ $y=1$
- $x$가 양의 무한대로 가면($x \rightarrow \infty$), 양의 무한대로 가까워짐
- $x$가 음의 무한대로 가면($x \rightarrow -\infty$), 0으로 가까워짐
- $x_1 > x_2$이면 $\exp{x_1} > \exp{x_2}$
- $ e^{x_1} e^{x_2} = e^{x_1 + x_2}$
- $ y = a^x = (e^{\log a})^x = e^{x \log a} $

```python
# 오일러수 표현하기
np.e
```

```python
# 지수 함수 구하기
x = 10
np.exp(x)
```

```python
# 지수 함수 그래프
xx = np.linspace(-2, 2, 100)
yy = np.exp(xx)

plt.axhline(1, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.grid(True)

plt.plot(xx, yy)
plt.show()
```

**3) 로그 함수 (log function)**

[1] 정의 : 지수함수의 역함수, 정확히는 자연 로그(natural log)함수라고 함

[2] 수식 : 밑 $e$는 보통 생략

$$ y = \log_e x\;\;\rightarrow\;\; y = \log x$$

[3] 특징

- $x$가 양수여야 함, 0이나 음수면 정의되지 않음
- $x > 1$ 이면 양수, $0 < x < 1$ 이면 음수
- $x=1\;\;\rightarrow\;\;y=0$
- $x_1 > x_2$이면 $\log{x_1} > \log{x_2}$

[4] 성질

- 로그 함수는 곱하기를 더하기로 변환

$$ \log{(x_1 \cdot x_2)} = \log{x_1} + \log{x_2} $$

- 어떤 함수에 로그를 적용해도 함수의 최고점, 최저점의 위치는 변하지 않음

$$ \arg\max_x f(x) = \arg\max_x \log f(x)  $$

```python
def f(x):
    return x**3 - 12*x + 20 * np.sin(x) + 7
```

```python
x = np.linspace(-4, 4, 300)
y = f(x)

plt.subplot(211)
plt.plot(x, y)
plt.axhline(1, c='r', ls="--")
plt.title("f(x)")
plt.grid(True)

plt.subplot(212)
plt.plot(x, np.log(y))
plt.axhline(0, c='r', ls="--")
plt.title("log f(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

- 0부터 1사이의 작은 값을 확대시켜 보여줌

```python
np.random.seed(0)
x = np.random.rand(5)
x = x / x.sum()

plt.subplot(211)
plt.bar(range(1, 6), x) # 일반적인 좌표 표시
plt.ylim(0, 1)

plt.subplot(212)
plt.bar(range(1, 6), np.log(x)) # 로그로 표현하면 값이 확대되어 보여짐

plt.show()
```

### 6. 다변수 함수 (Multivariate function)

1) 정의 : 복수의 입력 변수를 가질 수 있음

2) 수식 : $ z = f(x, y) $

3) 표현 방법

[1] 서피스 플롯 (surface plot)

```python
# 다변수 함수 - 서피스 플롯 그리기
def f(x, y):
    return 2 * x**2 + 6 * x * y + 7 * y**2 - 26 * x - 54 * y + 107
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
fig.gca(projection='3d').plot_surface(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Surface Plot")
plt.grid(True)

plt.show()
```

[2] 컨투어 플롯 (Contour plot)

```python
CS = plt.contour(X, Y, Z, levels=np.arange(0, 300, 20))

plt.clabel(CS, fmt="%d")
plt.title('Contour Plot')
plt.grid(True)

plt.show()
```

### 7. 다변수 다출력 함수 (Softmax function)

1) 정의 : 입력 변수 뿐 아니라 출력 변수도 여러개인 함수

2) 특징

- 모든 원소는 0와 1사이의 값을 가짐
- 모든 원소의 합은 1
- 입력 원소들의 크기 순서와 출력 원소들의 크기 순서가 같음

3) 예시 (3차원 벡터)

$$ 
\begin{bmatrix}
y_1 \\ y_2 \\ y_3
\end{bmatrix}
=h(x_1, x_2, x_3) = 
\begin{bmatrix}
\dfrac{\exp(w_1x_1)}{\exp(w_1x_1) + \exp(w_2x_2) + \exp(w_3x_3)} \\
\dfrac{\exp(w_2x_2)}{\exp(w_1x_1) + \exp(w_2x_2) + \exp(w_3x_3)} \\
\dfrac{\exp(w_3x_3)}{\exp(w_1x_1) + \exp(w_2x_2) + \exp(w_3x_3)} \\
\end{bmatrix}
$$

```python
# softmax 함수 구하기
def softmax(x, w):  # x는 1차원 배열, w는 가중치 벡터
    e = np.exp(w * x)
    return np.exp(w * x) / e.sum()

x = [2.0, 1.0, 0.5]
y = softmax(x, np.ones(3))
y
```

```python
np.sum(y)
```

```python
softmax(x, 4 * np.ones(3)) # 가중치가 커지면 최대값과 최소갑 차이가 더 벌어짐
```
