### 1. 확률 분포 함수

#### 1)  학습의 이유

- 통계 분석 과정에서는 수집된 데이터들이 특정 확률 분포 함수를 따른다는 가정을 통해 이루어지는 경우가 대부분
- 확률 분포 함수를 알게 되면, 구하고자 하는 사건에 대한 확률과 각종 정보의 유추가 가능

#### 2) 학습 내용

(1) 확률 변수의 정의 - 확률 변수가 어떤 상황에 대해 정의하고 있는지

(2) 확률 변수의 범위 - 확률 변수가 취할 수 있는 값의 범위는 어떻게 되는지

(3) 확률 분포함수의 형태 - 확률 분포 함수가 어떤 식으로 이루어져 있는지

(4) 확률 분포 함수를 통한 확률의 계산 방법

(5) 확률 분포함수의 요약값 - 확률 변수의 기대값과 분산은 어떻게 계산되는지

(6) 확률 분포함수의 모수 - 확률 분포함수의 모양, 형태 및 특성을 결정하는 값

2) 확률 밀도 함수 (PDF)

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 확률 밀도 함수 만들기
xx = np.linspace(-8, 8, 100) # -8에서 8사이에 100개의 데이터 생성
pdf = rv.pdf(xx) # 데이터를 pdf로 정의

plt.grid(True)
plt.plot(xx, pdf)
plt.show()
```

3) 누적 분포 함수 (CDF)

```python
# 누적 분포함수 만들기
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)

plt.grid(True)
plt.plot(xx, cdf)
plt.show()
```
