###  연습문제 2

- 입력 데이터는 10개 이상으로 설정한다


#### 1. 입력 데이터 : 제품 종류, 제품 가격, 면적, 친절도, 접근성(상권과 입지), 주차공간, 청결도, 이벤트, 경쟁 마트 존재여부, 편의성


- 제품 종류 : 제품 종류의 수를 의미한다.
- 제품 가격 : 제품 가격을 의미한다.
- 면적 : 층수를 포함한 전체 면적을 의미한다.
- 친절도 : 직원이 고객을 응대하는 친절도를 의미한다. (설문조사로 1~5점 부여)
- 접근성 : 주위에 버스 정류장, 지하철 등 대중교통이 편리한 부분과 주위 편의 시설(공원, 영화관 등)의 접근성을 의미한다. (주위 200m, 400m, 600m, 800m, 1000m이상 순으로 5~1점 부여)
- 주차공간 : 주차할 수 있는 차량이 몇대가 되는지를 의미한다.
- 청결도 : 마트 시설(화장실, 진열대 등)의 청결도를 의미한다. (1~5점 부여)
- 이벤트 : 1+1행사나 가격할인 행사등 분기별로 얼마나 많은 이벤트를 시행하는지를 의미한다.
- 경쟁마트 존재여부 : 5km주위에 경쟁 대형마트가 있는지를 의미한다.
- 편의성 : 고객들이 물건을 찾기 쉽게 구조가 되어있는지를 의미한다. (설문조사로 1~5점 부여)


#### 2. 입력 데이터 : 득점, 리바운드, 어시스트, 스틸, 블록, 자유투 성공률, 2점슛 성공률, 3점슛 성공률, 부상일수, 출전 시간, 턴오버, 드래프트 순위


- 득점: 경기당 평균 득점 수를 의미한다. 
- 리바운드 : 경기당 평균 리바운드 수를 의미한다.
- 어시스트 : 경기당 평균 어시스트 수를 의미한다.
- 스틸 : 경기당 평균 스틸 수를 의미한다.
- 블록 : 경기당 평균 블록을 의미한다.
- 자유투 성공률 : 시즌 중 전체 자유투 시도 중 성공한 개수의 비율을 의미한다.
- 2점슛 성공률 : 시즌 중 전체 2점슛 시도 중 성공한 개수의 비율을 의미한다.
- 3점슛 성공률 : 시즌 중 전체 3점슛 시도 중 성공한 개수의 비율을 의미한다.
- 부상일수 : 지금까지 선수 생활 기간 중 부상일수의 비율을 의미한다.
- 출전시간 : 경기당 평균 출전시간을 의미한다.
- 턴오버 : 경기당 턴오버 수를 의미한다.
- 드래프트 순위 : 드래프트에 뽑힌 순서를 의미한다.


#### 3. 입력 데이터 : 노인복지 시설수, 음주율, 흡연율, 이혼율, 스트레스 인지율, 우울감, 기초수급자 비율, 사회복지예산, 1인가구, 삶의질


- 노인복지 시설수 : 광역시별 노인 인구당 노인 복지 시설수
- 음주율 :  광역시별 노인 인구당 음주 경험 있는 노인 인구 (음주율: 소주 기준 30일간 1회 평균 남자 5잔 이상, 여자 3잔 이상 음주 비율)
- 흡연율 : 광역시별 노인 인구당 흡연 경험 있는 노인 인구 (흡연율: 현재 담배를 피우는 비율)
- 이혼율 : 광역시별 노인 인구당 이혼한 노인 인구
- 스트레스 인지율 : 평상시 스트레스를 '많이' 느끼는 노인의 비율
- 우울감 : 광역시별 노인 인구당 우울감을 느끼는 노인 인구
- 기초수급자 비율 : 광역시구별 노인 인구당 기초수급자 수
- 사회복지예산 : 광역시구 전체 예산 중 사회복지예산이 차지하는 비율
- 1인가구 : 광역시구별 노인 인구당 1인 가구 수
- EQ-5D : 광역시구별 노인 인구당 운동능력, 자기관리, 일상활동, 통증·불편감, 불안·우울감 등의 삶의 만족에 관한 점수

###  연습문제 3

#### 1. 대형마트 매출 상승 예측분석


- 제품 종류 : 제품의 수가 많을 수록 고객의 수요를 충족할 가능성이 높아 매출이 높을 것
- 제품 가격 : 가격이 높을수록 수요가 낮아져 매출이 낮을 것.
- 면적 : 면적이 크면 그만큼 다양한 제품을 전시할 수 있기 때문에 매출이 높을할 것
- 친절도 : 친절도가 높을수록 충성고객이 증가하기 때문이 매출이 높을 것
- 접근성 : 접근성이 좋다는 것은 그만큼 방분 빈도가 높을 가능성이 높기 때문에 매출이 높을 것
- 주차공간 : 주차공간이 넓을수록 주차에 대한 거부감이 줄어들 가능성이 높기 때문에 매출이 높을 것
- 청결도 : 청결도가 높을 수록 고객들이 마트에 대한 이미지가 좋아질 것이기 때문에 매출이 높을 것
- 이벤트 : 이벤트가 주로 가격관련 이벤트가 많은데 이벤트가 많을수록 매출이 높을 것
- 경쟁마트 존재여부 : 주위에 경쟁마트가 있으면 그만큼 고객을 빼앗길 가능성이 높기 때문에 매출이 낮을 것
- 편의성 : 구조가 쉽게 되있으면 마트에 대한 거부감이 줄어들기 때문에 매출이 높을 것


- 정리: 제품종류, 면면적, 친절도, 접근성, 주차공간, 청결도, 이벤트, 편의성에 대한 변수는 (+)요인으로 두고 가격과 경쟁마트 존재여부를 (-)로 하는 식을 도출해 내면 매출상승을  예측할 수 있을 것이다.


#### 2. KBL 농구선수 연봉 예측

- 득점: 득점역이 높을수록 공격력을 증가시키기 때문에 연봉이 높을 것
- 리바운드 : 리바운드는 수비력을 증가시키기 때문에 연봉이 높을 것
- 어시스트 : 어시스트가 높을수록 공격력을 증가시키기 때문에 연봉이 높을 것
- 스틸 : 스틸이 높을수록 수비력을 증가시키기 때문에 연봉이 높을 것
- 블록 : 블록이 높을수록 수비력을 증가시키기 때문에 연봉이 높을 것
- 자유투 성공률 : 자유투가 높을수록 공격력을 증가시키기 때문에 연봉이 높을 것
- 2점슛 성공률 : 2점슛이 높을수록 공격력을 증가시키기 때문에 연봉이 높을 것.
- 3점슛 성공률 : 3점슛이 높을수록 공격력을 증가시키기 때문에 연봉이 높을 것
- 부상일수 : 부상일수가 높을수록 경기에 출전하는 일수가 적기 때문에 연봉이 낮을 것
- 출전시간 : 출전시간이 높을수록 팀에 기여하는 빈도가 높은 선수이기 때문에 연봉이 높을 것
- 턴오버 : 턴오버가 높을수록 공격력을 감소시키기 때문에 연봉이 낮을 것
- 드래프트 순위 : 드래프트 순위가 높다는건 선수의 실력이 검증되었다는 것을 의미하기 때문에 연봉이 높을 것


- 정리: 득점, 리바운드, 어시스트, 스틸, 블록, 자유투 성공률, 2점슛 성공률, 3점슛 성공률, 출전시간, 드래프트 순위에 대한 변수는 (+)요인으로 두고 부상일수, 턴오버에 여부를 (-)로 하는 식을 도출해 내면 연봉을 예측할 수 있을것이다. 


#### 3. 광역시구별 우리나라 노인 자살율 예측


- 노인복지 시설수 : 복지시설이 높을수록 노인분들이 다른 사람과의 접촉도 많아지고 사회적 존재로 인지할 가능성이 높기 때문에 자살율이 낮을 것
- 음주율 : 음주율이 높을수록 극단적인 생각과 판단이 많아지기 때문에 자살율이 높을것 
- 흡연율 : 흡연을 하는 목적 중 하나가 스트레스 원인이 있기 때문에 흡연율이 높을 수록 자살률이 높을 것
- 이혼율 : 이혼을 통해 가정의 불화가 생기고 부정적 영향을 미칠 수 있기 때문에 이혼율이 높을수록 자살율이 높을 것
- 스트레스 인지율 :스트레스는 극단적 판단을 불러올 수 있기 때문에 스트레스 인지율이 높으면 자살율이 높을 것
- 우울감 : 우울감이 높으면 사회에 대한 불만이 생길 수 있기 때문에 우울감이 높으면 자살율이 높을 것
- 기초수급자 비율 : 기초수급자가 많으면 그만큼 경제적으로 힘든 사람이 많은것을 의미하기 때문에 자살율이 높을 것
- 사회복지예산 : 사회복지예산이 증가하면 노인들 복지에 긍정적 영향을 미치기 때문에 자살율이 낮을 것
- 1인가구 : 1인 가구는 사회실제감 결여를 느낄 가능성이 높기 때문에 1인가구가 많을수록 자살율이 높을 것
- EQ-5D : 삶의 질이 높으면 현재 삶에 대한 만족도가 크기 때문에 자살율이 낮을 것


- 정리: 음주율, 흡연율, 이혼율, 스트레스 인지율 우울감, 기초수급자 비율, 1인가구에 대한 변수는 (+)요인으로 두고 노인복지 시설수, 사회복지예산, EQ-5D에 대한 변수는 (-)로 하는 식을 도출해내면 노인 자살율을 예측할 수 있을 것이다
