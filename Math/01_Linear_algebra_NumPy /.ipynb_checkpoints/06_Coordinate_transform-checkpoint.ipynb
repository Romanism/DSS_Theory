### 좌표와 변환

### 1. 벡터의 선형 종속과 선형 독립

1) 선형 종속(Linearly dependent)

[1] 선형 조합 $ c_1 x_1 + c_2 x_2 + \cdots + c_N x_N $ = 0(영벡터) 에서

[2] 모두 0이 아닌 스칼라값 $c_1, c_2, \ldots, c_N$이 존재할 때 선형 종속이라 함

ex)

$$ 
x_1 = \begin{bmatrix}1 \\ 2\end{bmatrix} ,\;\;
x_2 = \begin{bmatrix}3 \\ 3\end{bmatrix} \;\;
x_3 = \begin{bmatrix}10 \\ 14\end{bmatrix} \;\;
$$

이는 다음과 같은 식으로 증명할 수 있다.

$$ 
2x_1  + x_2 - \frac{1}{2} x_3 = 0
$$

따라서 벡터 $x_1, x_2, x_3$는 선형 종속

```python
# 선형 종속 구하기 (NumPy)
x1 = np.array([1, 2])
x2 = np.array([3, 3])
x3  = np.array([10, 14])
2 * x1 + x2 - 0.5 * x3
```

2) 선형 독립(Linearly independent)

[1] 선형 조합 $ c_1 x_1 + c_2 x_2 + \cdots + c_N x_N $ = 0(영벡터) 에서 

[2] 모두 0이 아닌 스칼라값 $c_1, c_2, \ldots, c_N$이 존재하지 않을 때 선형 독립이라 함

3) 선형 조합

선형 조합을 알아내는 문제는 선형 연립 방정식을 푸는 문제와 같음

$$ 
\begin{eqnarray}
c_1 x_1 + \cdots + c_N x_N
&=&
\begin{bmatrix} x_1 & x_2 & \cdots & x_N \end{bmatrix} 
\begin{bmatrix} c_1 \\ c_2 \\ \vdots \\ c_N \end{bmatrix}
&=& X c
= 0
\end{eqnarray}
$$

[1] X의 역행렬이 존재하면 $ c = X^{-1}0 = 0 $이라 0벡터가 유일한 해 $\rightarrow$ 선형독립

[2] 행의 갯수 < 열의 갯수, 그 외의 이유로 0 벡터가 아닌 해 $c$가 존재하면 선형 종속

### 2. 랭크

1) 정의

- 열 랭크 (column rank) : 행렬의 열 벡터 중 서로 독립인 열 벡터의 최대 갯수
- 행 랭크 (row rank) : 행 벡터 중 서로 독립인 행 벡터의 최대 갯수
- 풀 랭크 (full rank) : 랭크가 행의 갯수와 열의 갯수 중 작은 값과 같은 경우

다음 행렬 $A$의 랭크는 2

$$ 
A = \begin{bmatrix} 1 & 5 & 6 \\ 2 & 6 & 8 \\ 3 & 11 & 14\\ 1 & 4 & 5 \end{bmatrix} 
$$

다음 행렬 $B$의 랭크는 3

$$ 
B = \begin{bmatrix} 1 & 5 & 6 \\ 2 & 6 & 8 \\ 3 & 11 & 14\\ 1 & 4 & 8 \end{bmatrix} 
$$

2) 성질 - 행 랭크와 열 랭크는 항상 같음

```python
# 랭크 구하기 (NumPy)
A = np.array([[1, 5, 6], [2, 6, 8], [3, 11, 14], [1, 4, 5]])
np.linalg.matrix_rank(A)
```

```python
B = np.array([[1, 5, 6], [2, 6, 8], [3, 11, 14], [1, 4, 8]])
np.linalg.matrix_rank(B)
```

### 3. 벡터 공간 (vector space) 과 기저벡터 (basis vector)

1) 정의

- 벡터 N개가 서로 선형 독립일 경우에 정의
- 벡터 공간 : 선형 독립인 벡터들을 조합하여 만들어지는 모든 벡터의 집합
- 기저 벡터 : 벡터 공간안에 있는 벡터들

ex) 2차원 벡터 공간의 기저 벡터

$$ 
a_1 = \begin{bmatrix}1 \\ 3\end{bmatrix} ,\;\;
a_2 = \begin{bmatrix}0 \\ 1\end{bmatrix} \;\;
$$

2) 표준 기저 벡터 (Standard basis vector)

$$ 
e_1 = \begin{bmatrix} 1 \\ 0 \\ \vdots \\ 0 \end{bmatrix}, \;\;
e_2 = \begin{bmatrix} 0 \\ 1 \\ \vdots \\ 0 \end{bmatrix}, \;\; 
\cdots,  \;\;
e_N = \begin{bmatrix} 0 \\ 0 \\ \vdots \\ 1 \end{bmatrix}
$$

### 4. 좌표 표현 (Coordinate representation)

1) 정의 : 기저 벡터를 선형 조합하여 해당 벡터를 나타냈을 때 선형조합의 가중치

$$ x = r_1 e_1 + r_2 e_2 \;\;에서 $$

$$ \begin{bmatrix} r_1 \\ r_2 \end{bmatrix} $$

를 벡터 $x$의  기저 벡터 $e_1$, $e_2$에 대한 좌표라고 함

2) 성질 : 표준 기저 벡터에 대한 좌표는 원래 벡터와 같음

### 5. 변환 행렬

기존의 기저 벡터  $e_1, e_2$와 새로운 기저 벡터 $g_1, g_2$간 다음과 같은 관계가 성립한다고 가정

$$
g_1 = \dfrac{1}{\sqrt{2}} e_1 + \dfrac{1}{\sqrt{2}} e_2
$$

$$
g_2 = -\dfrac{1}{\sqrt{2}} e_1 + \dfrac{1}{\sqrt{2}} e_2
$$

이 때 $e_1$, $e_2$에 대한 $g_1$, $g_2$의 좌표 벡터는 다음처럼 열 벡터로 나타낼 수 있음

$$
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} 
\end{bmatrix}, \;\;
\begin{bmatrix} 
-\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
$$

행렬 $A$은 이 열 벡터들을 합친 행렬 이를 변환행렬이라 함

$$ A =
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & -\dfrac{1}{\sqrt{2}} \\
\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
$$

### 6. 좌표 변환 (Coordinate transform)

1) 정의 : 새로운 기저 벡터에 대해 좌표를 계산하는 것

2) 과정

[1] $a$의 기저 벡터 $e_1, e_2$ 좌표를 새로운 기저 벡터 $g_1, g_2$에 대한 좌표를 변환하고자 함

[2] 새로운 기저 벡터에 대한 좌표 값이 가리키는 실제 위치는 원래의 벡터가 가리키는 실제 위치과 같아야 함

$$ a = a_1e_1 + a_2e_2 = {a'}_1 g_1 + {a'}_2 g_2 $$

[3] 이 식에서 ${a'}_1, {a'}_2$가 새로운 기저 벡터에 대한 좌표

$$ {a'}_1 g_1 + {a'}_2 g_2 =
\begin{bmatrix} g_1 & g_2 \end{bmatrix} 
\begin{bmatrix} {a'}_1 \\ {a'}_2 \end{bmatrix}
= \begin{bmatrix} g_1 & g_2 \end{bmatrix} a' = A a'
$$ 

[4] $\; Aa' = a \; \rightarrow a' = A^{-1}a = Ta $

[5] A의 역행렬 T=A−1을 변환 행렬(transform matrix)이라고 함

원점을 이동하지 않는 선형 좌표 변환은 행렬을 곱하여 구할 수 있음

ex) 

벡터 $a$의 표준 기저 벡터에 대한 좌표

$$
a = 2 e_1 + 2 e_2= \begin{bmatrix}2 \\ 2\end{bmatrix} 
$$

표준 기저 벡터에 대한 새로운 기저 벡터의 좌표

$$ 
g_1 = \begin{bmatrix} \dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix} ,\;\;
g_2 = \begin{bmatrix} -\dfrac{1}{\sqrt{2}} \\ \dfrac{1}{\sqrt{2}} \end{bmatrix} \;\;
$$

새로운 기저 벡터에 대한 벡터 $a$의 좌표

$$ 
a' = A^{-1}a = 
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & -\dfrac{1}{\sqrt{2}} \\
\dfrac{1}{\sqrt{2}} &  \dfrac{1}{\sqrt{2}} 
\end{bmatrix}^{-1}
\begin{bmatrix}2 \\ 2\end{bmatrix}
=
\begin{bmatrix} 
\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} \\
-\dfrac{1}{\sqrt{2}} & \dfrac{1}{\sqrt{2}} 
\end{bmatrix}
\begin{bmatrix}2 \\ 2\end{bmatrix} 
= \begin{bmatrix}2\sqrt{2}\\0\end{bmatrix} 
$$

### 7. 이미지 변환

- 기저 벡터에 대한 좌표 변환을 응용하면 이미지를 자유롭게 변환 가능
- 이미지를 이루는 픽셀을 새로운 좌표로 아동시키는 아이디어

$$ 
e_1 = \begin{bmatrix} 0 \\ -1 \end{bmatrix} ,\;\;
e_2 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

```python
import scipy as sp
import scipy.misc

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
# 
f = sp.misc.face(gray=True)
plt.imshow(f, cmap=mpl.cm.bone)
plt.grid(False)
plt.show()
```

- 인수로는 이미지 데이터와 변환 행렬의 역행렬을 받음
- 파이썬 이미지에선 표준 기저벡터를 사용, 원점이 왼쪽 상단에 있음

```python
import scipy.ndimage

Ainv = np.array([[ 0.70710678, -0.70710678],[ 0.70710678,  0.70710678]])
f1 = sp.ndimage.affine_transform(f, Ainv) # A의 역행렬 만큼 잘려서 이미지 출력
plt.imshow(f1, cmap=mpl.cm.bone)
plt.grid(False)
plt.show()
```
